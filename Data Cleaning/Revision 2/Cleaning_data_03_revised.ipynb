```python
place_ids = []
operation_hours = []
for item in data:
    result = item["result"]
    place_id = result.get("place_id")
    place_ids.append(place_id)

    opening_hours = result.get('opening_hours')
    if opening_hours is not None:
        weekday_text = opening_hours.get('weekday_text')
        if weekday_text is not None:
            operation_hours.append(weekday_text)
        else:
            operation_hours.append(np.nan)
    else:
        operation_hours.append(np.nan)
```

```python
def split_hours(row):
    if pd.isna(row):
        return [np.nan] * 7

    days = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']
    hours = [np.nan] * 7

    for day in row.split(','):
        parts = day.split(':')
        if len(parts) == 2:
            day_name = parts[0].strip()
            day_index = days.index(day_name)
            hours[day_index] = parts[1].strip()

    return hours

OperationHours[['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']] = pd.DataFrame(OperationHours['opening_hours_str'].apply(split_hours).tolist())
```

```python
def split_hours(row):
    if pd.isna(row):
        return [np.nan] * 14

    days = ["Senin", "Selasa", "Rabu", "Kamis", "Jumat", "Sabtu", "Minggu"]
    hours = [np.nan] * 14


    for day in row.split(','):
        parts = day.split(':')
        if len(parts) == 2:
            day_name = parts[0].strip()
            day_index = days.index(day_name)
            hour_range = parts[1].strip()

            if hour_range == 'Buka 24 jam':
                hours[day_index * 2] = '00.00'
                hours[day_index * 2 + 1] = '23.59'
            elif hour_range == 'Tutup':
                hours[day_index * 2] = 'Close'
                hours[day_index * 2 + 1] = 'Close'
            else:
                try:
                    open_time, close_time = hour_range.split('–')
                    hours[day_index * 2] = open_time.strip()
                    hours[day_index * 2 + 1] = close_time.strip()
                except ValueError:
                    hours[day_index * 2] = np.nan
                    hours[day_index * 2 + 1] = np.nan

    return hours
```

```python
new_columns = ['Monday_Open', 'Monday_Close', 'Tuesday_Open', 'Tuesday_Close', 'Wednesday_Open', 'Wednesday_Close', 'Thursday_Open', 'Thursday_Close', 'Friday_Open', 'Friday_Close', 'Saturday_Open', 'Saturday_Close', 'Sunday_Open', 'Sunday_Close']

OperationHours[new_columns] = pd.DataFrame(OperationHours['opening_hours_str'].apply(split_hours).tolist())
```

## Types

```python
place_ids = []
types = []
for item in data:
    result = item["result"]
    place_id = result.get("place_id")
    place_ids.append(place_id)

    categories = result.get('types')
    types.append(categories)
```

```python
categories = pd.DataFrame({'place_id' : place_ids, 'types' : types})
```

```python
categories.head()
```

```python
place_types = ['bar', 'cafe', 'restaurant']
```

```python
# Create a new DataFrame with dummy variables for each type
dummy_df = pd.get_dummies(categories['types'].explode())

# Reindex the columns to include all possible types
dummy_df = dummy_df.reindex(columns=place_types, fill_value=0)

# Sum the dummy variables for each type to get the final 1 or 0 values
df_types = dummy_df.groupby(level=0).sum()

# Concatenate the original DataFrame with the types DataFrame
categories = pd.concat([categories, df_types], axis=1)
```
